```python
left = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})

result = pd.merge(left, right, on = ['key1', 'key2'])
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'left')
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'right')
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'outer')
result
```

```python
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'right', indicator = True)
result
```

```python
pop_seoul = pd.read_csv('서울인구통계.txt', sep = '\t', engine = 'python', thousands = ',', encoding = 'utf-8')
pop_seoul.head()
```
