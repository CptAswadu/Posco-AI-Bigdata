# 자동차 모델과 연비

```python
df = ds_mycars[['model', 'mpg']]
df.head()
```

```python
count = df['model'].value_counts().sort_index()
cumcnt = np.cumsum(count)
percent = count/sum(count) * 100
cumpct = np.cumsum(percent)
count_data = pd.DataFrame({'Count': count, 'Comcnt': cumcnt, 'Percent': percent, 'Cumpct': cumpct})
count_data.columns.name = 'model'
count_data
```

```python
count2 = df['mpg'].value_counts().sort_index()
cumcnt2 = np.cumsum(count2)
percent2 = count2/sum(count) * 100
cumpct2 = np.cumsum(percent2)
count_data2 = pd.DataFrame({'Count': count2, 'Comcnt': cumcnt2, 'Percent': percent2, 'Cumpct': cumpct2})
count_data2.columns.name = 'mpg'
count_data2
```

# 대한민국 남성 키: 
## 나의 키가 180cm일 경우 나는 몇% 범위에 위치할까요?

```python
mu = 175
sigma = 5
x = 180
prob = stats.norm.cdf(x, mu, sigma)
print("{0}cm 위치 비율: {1:.1f}%".format(x, (1-prob)*100))
```

### 자동차 1리터당 주행거리 평균  12km, 표준편차 3km 정규분포
## 1. 9km이상 15km이하 달릴 확률은?

```python
mu = 12
sigma = 3
x1 = 15
x2 = 9
prob1 = stats.norm.cdf(x1, mu, sigma)
prob2 = stats.norm.cdf(x2, mu, sigma)
prob = prob1 - prob2
print("{0}km 이상 {1}km 이하를 달릴 확률은 {2:.2f}%에 해당함".format(x2, x1, (prob)))
```

### 자동차 1리터당 주행거리 평균  12km, 표준편차 3km 정규분포
## 2. 15km이상 달릴 확률은?

```python
mu = 12
sigma = 3
x1 = 15
prob1 = 1 - stats.norm.cdf(x1, mu, sigma)
print("{0}km 이상 달릴 확률은 {1:.2f}%에 해당함".format(x1, prob1))
```
